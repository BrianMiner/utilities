```python
toexport=pd.DataFrame(pred_new)
toexport['id']=x_test.id
toexport.columns=['new_pred','id']
```

```python

def quintile_classify(x):
    if x >= 0.003812:
        quintile=1
    elif x < 0.003812 and x >= 0.003403:
        quintile=2
    elif x < 0.003403 and x >= 0.003109:
        quintile=3
    elif x < 0.003109 and x >= 0.002821:
        quintile=4
    elif x < 0.002821:
        quintile=5
    else:
        quintile =99
    return(quintile)



toexport['quintile']=toexport.new_pred.apply(quintile_classify)
```
